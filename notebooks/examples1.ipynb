## Example 1.15 $\mathrm{H}_{0}$: $\mu = 1$ *vs* $\mathrm{H}_{1}$: $\mu \gt 1$

```r
xdata <- c(0.5, 0.9, 4.5, 3.4, 1.0,
2.7, 1.1, 1.9, 0.0, 0.0,
4.2, 2.1, 0.0, 2.0, 3.4,
3.4, 2.5, 0.9, 5.1, 2.4)
n <- length(xdata)
xbar <- mean(xdata)
s <- sd(xdata)
se_xbar <- s/sqrt(n)
mu0 <- 1
#  test stat
t <- (xbar-mu0)/se_xbar
# pvalue 
1-pt(t,n-1)
```

## class example lm

```r
x = c(1,2,3,4,5)
y = c(1,1,2,2,4)
fit = lm(y ~ x)
summary(fit)
y_hat = -0.1 + 0.7*x
plot(x, y, ylim=c(0, 5))
lines(x, y_hat)
sigma(fit)
```
